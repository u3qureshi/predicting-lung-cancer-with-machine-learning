```python
# VISUALIZING THE DATA PER AGE IN THE DATASET OF THOSE WHO HAVE LUNG CANCER
# First create a lung cancer column for true value counts
# DUMMY VARIABLE CODING
series = df['LUNG_CANCER']
series = series.replace(to_replace = ['YES'], value = [1])
series = series.replace(to_replace = ['NO'], value = [0])

# Insert list into the dataframe
df['LUNG_CANCER_INT'] = series.values

# DISPLAY THE DATA GROUPED BY AGE IN THE DATASET OF THOSE WHO ARE POSITIVE OF LUNG CANCER
dfByAge = df.groupby('AGE').agg({'LUNG_CANCER_INT' : 'sum'})
dfByAge

```

```python
# First create a GENDER cancer column for true value counts, 0=FEMALE AND 1=MALE
# DUMMY VARIABLE CODING
series = df['GENDER']
series = series.replace(to_replace = ['M'], value = [1])
series = series.replace(to_replace = ['F'], value = [0])

# Insert list into the dataframe
df['GENDER_MALE'] = series.values

# NEW COLUMN GENDER_MALE ADDED TO THE DATAFRAME FOR THE PURPOSE OF STANDARDIZING AND PLOTTING THE GENDER DATA ALONG WITH THE REST ONTO A LOGISTIC REGRESSION PLOT
df.head()
```

```python
# NOW GROUP BY GENDER AND DISPLAY THOSE POSITIVE WITH LUNG CANCER
dfByGender = df.groupby('GENDER').agg({'LUNG_CANCER_INT' : 'sum'})
dfByGender

# It is evident that age and gender have no serious correlation to lung cancer, although age may play a slight factor because this dataset has very few to none individuals below 40 years of age
# It can also be seen that there are more males with lung cancer
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df.drop(['LUNG_CANCER','LUNG_CANCER_INT','GENDER'],axis = 1))
scaledVariables = scaler.transform(df.drop(['LUNG_CANCER','LUNG_CANCER_INT','GENDER'],axis = 1))
dfScaled = pd.DataFrame(scaledVariables,columns = ['AGE', 'SMOKING', 'YELLOW FINGERS', 'ANXIETY', 'PEER PRESSURE',
       'CHRONIC DISEASE', 'FATIGUE', 'ALLERGY', 'WHEEZING',
       'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH',
       'SWALLOWING DIFFICULTY', 'CHEST PAIN',  'GENDER'])
dfScaled

X = dfScaled
Y = df['LUNG_CANCER_INT']

# First we split the data with 80% training data
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, random_state=0, train_size=0.8)

lr = LogisticRegression()
lr.fit(Xtrain, Ytrain)
predictions = lr.predict(Xtest)
print(classification_report(Ytest,predictions))
print("The accuracy score for the logistic regression model is " + str(round(accuracy_score(Ytest, predictions)*100, 2)) + "%")
```

```python
from sklearn.metrics import confusion_matrix 
# This confusion matrix is a table which is describing the performance this logistic regression classification model. This logistic regression model was successful in predicting positive lung cancer cases 51 out of 52 times, which is very good. This logistic regression model also successfully predicted the negative lung cancer cases 5 times out of 10. This is not as good as the positive lung cancer predictions but this could be due to the significantly lower amount of data on negative lung cancer. Regardless of the mediocrity of the prediction of the negative lung cancer cases, this confusion matrix shows a high precision of the model.
cm = confusion_matrix(Ytest,predictions)
sns.heatmap(cm, annot=True, cmap="Greens")
plt.title('Confusion matrix of the logistic regression model')
plt.show()
```

```python
# Prediction of lung cancer using Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(random_state=0)
rfc.fit(Xtrain, Ytrain)
prediction = rfc.predict(Xtest)
print("The accuracy score for the random forest classifier model is " + str(round(accuracy_score(Ytest, prediction)*100, 2)) + "%")
```

```python
cm = confusion_matrix(Ytest,prediction)
sns.heatmap(cm, annot=True, cmap="Greens")
plt.title('Confusion matrix of the random forest classifier model')
plt.show()

# As can be seen, this random forest classifier model does equally well when predicting the positive lung cancer cases and it performs much better with predicting the negative lung cancer cases.
```
